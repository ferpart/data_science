```python
import numpy as np
from datascience import *
```

```python
choices = make_array("first goat", "second goat", "car")
length = 100
```

```python
first_choice = np.random.choice(choices, length)
final_choices = []
for i in range(length):
    actual_choice = choices.tolist()
    index = np.where(choices == first_choice[i])
    #print(index[0])
    actual_choice.pop(index[0][0])
    final_choices.append(actual_choice)

exposed = []
remaining = []
for i in range(length):
    if "first goat" in final_choices[i]:
        final_choices[i].remove("first goat")
        exposed.append("first goat")
    elif "second goat" in final_choices[i]:
        final_choices[i].remove("second goat")
        exposed.append("second goat")

for i in range(length):
    remaining.append(final_choices[i][0])

remaining

choice_table = Table().with_columns("Trial", np.arange(length)+1, "Guess", first_choice, "Exposed", exposed, "Remaining", remaining)
choice_table.show()
```
